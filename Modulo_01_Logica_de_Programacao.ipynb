## Módulo 1 - Introdução à Logica de Programação

### Conceitos Básicos

#### Algoritmos

Um algoritmo é uma lista passo-a-passo de instruções que, caso seguido exatamente, atinge um objetivo determinado.
Seria algo como uma "receita" para executarmos uma tarefa ou resolver algum problema. E como toda receita, um algoritmo
também deve ser finito. Se seguirmos uma receita de bolo corretamente, conseguiremos fazer o bolo. O algoritmo pode ser
representado por um código, um pseudo-código ou um fluxograma.

- *Código*: É a implementação concreta do algoritmo contendo comandos de linguagem de programação;
- *Pseudo-Código*: Descrição de alto nível (com estrutura e operações abstratas) do algoritmo;
- *Fluxograma*: É uma representação gráfica de um algoritmo.

#### Operadores em Python:

- *Operadores Aritméticos*: Adição(+),  Subtração(-),  Multiplicação(*),  Divisão(/),  Divisão por inteiro(//), Resto da divisão(%), Exponenciação(**)

- *Operadores Relacionais*: Igual(==), Não igual(!=), Maior(>), Maior ou igual(>=), Menor(<), Menor ou igual(<=)

- *Operadores de Atribuição*: Atribuição(=), Adição e atribuição(+=), Subtração e atribuição(-=), Multiplicação e atribuição(*=), Divisão e atribuição(/=), Divisão por inteiro e atribuição(//=), Resto da divisão e atribuição(%=), Exponenciação e atribuição(**=)

- *Python Logical Operator*: “E” lógico(and), “Ou” lógico(or), Negação(not)

- *Operadores de Pertencimento*: “Pertence”(∈) ao conjunto(in), “Não pertence”(∉) ao conjunto(not in)

- *Operadores de identidade*: “É” a mesma coisa (is), “Não é” a mesma coisa (is not)

- *Operadores bit a bit*: “E” binário (&), “Ou” binário (|), “Ou exclusivo” binário (^), “Deslocamento para a esquerda”(<<), “Deslocamento para a direita” binário (>>), “Complemento” binário (~)

#### Variáveis:

As variáveis são utilizadas em um código Python para armazenar valores que queremos usar mais tarde. São espaços armazenados no computador que armazenam valores.

Existem algumas regras para a criação do nome das variáveis:

    1 – Não podem começar com números;
    2 – Não pode haver espaço no nome (se necessário utilizar _ );
    3 – Não é possível usar esses símbolos: “ “, <>, /, |, \, ( ), @, #, $, %, ^, &, *, ~, -, +, !
    4 – Não se podem usar palavras reservadas (que já são usadas pelo python) Ex: false, true, if, elife.

#### Decisões:

A lógica de decisão é usada para decidir algum fluxo do algoritmo baseado em alguma condição estabelecida.
Python possui 3 operadores de decisão: if, elife, else

- *if(se)*: O condicional if nos permite dizer ao computador ações com base em um determinado conjunto de resultados. 
Verbalmente, é como se disséssemos ao computador: “Ei, caso aconteça isso, execute essa ação”. O if é uma estrutura de 
condição que permite avaliar uma expressão e, de acordo com seu resultado, executar uma determinada ação.
    
- *elif(se)*: Adicionalmente, se existir mais de uma condição alternativa que precisa ser verificada, deve-se utilizar o elif para avaliar as expressões intermediárias antes de usar o else.

- *else(se não)*: O condicional else executa um comando caso a expressão if seja falsa.

#### Repetições

Em algumas situações é comum que uma mesma instrução (ou conjunto delas) precise ser executada várias vezes seguidas. 
Nesses casos, normalmente utilizamos um loop (ou laço de repetição) que permite executar o mesmo bloco de código 
enquanto uma condição é atendida. Em Python, os loops são codificados com as estruturas de repetição for e while.

- *For*: O laço for nos permite percorrer os itens de uma coleção e, para cada um deles, executar o bloco de código 
declarado no loop.

- *While(enquanto)*: O while é uma estrutura de repetição utilizada quando queremos que determinado bloco de código seja 
executado ENQUANTO (do inglês while) determinada condição for satisfeita. Em outras palavras: só saia da estrutura de 
repetição quando a condição não for mais satisfeita.
