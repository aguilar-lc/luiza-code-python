## Módulo 09 - Aplicando Segurança no Código

### Pilares de Segurança:
Os pilares são princípios que devem ser aplicados em todos os sistemas para torná-los seguros. Para isso temos a Tríplica que compoe os Pilares de Segurança (CID):

- Confidencialidade:
    Garantir que a informação não será divulgada ou disponibilizada para pessoas não autorizadas.
    Propriedade de manter sigilo.

- Integridade:
    Garantir que a informação não será modificada por pessoas não autorizadas.

- Disponibilidade
    Garantir que a informação estará disponível para as pessoas autorizadas.

### Autenticidade:
- Garantir que a pessoa é quem ela diz ser. Para isso, temos algumas formas de aferir a autenticidade:

    - Pelo o que a pessoa sabe: Senha/ Conhecimento/ Passado da pessoa
        - Vantagens: baixo custo
        - Desvantagens: se a informação for vazada, uma pessoa pode se passar por outra

    - Pelo o que a pessoa possui: Documento/ Token/ Coroa de ouro do rei
        - Vantagens: não pode ser facilmente reproduzido
        - Desvantagens: pode ser transferido de uma pessoa para outra

    - Pelo o que a pessoa é: Biometria facial/ Digital/ Personalidade
        - Vantagens: baixo custo, instranferível
        - Desvantagens: se a informação for vazada, os dados (biometria, digital, por exemplo) não pode ser alterada como
          em uma senha.

### Não Repúdio:
- Garantir que a pessoa não possa negar os seus atos.
- Está associado a integridade e origem do dado

### Autenticação e Autorização:
- Diferentes usuários podem precisar de diferentes permissões dentro do sistema. Exemplo: permissões dos usuários para leitura, escrita, etc.

    - Autenticação:
        - Garantir que a pessoa é quem ela diz ser.
        - através das formas de aferição.
        - Exemplo: apresentar o crachá para entrar em uma empresa

    - Autorização:
        - Conjunto de permissões disponíveis para uma pessoa
        - se uma pessoa pode ou não fazer algo ou estar em um lugar.
        - Exemplo: Mesmo com o crachá de uma empresa e entrando nela(autenticação) uma pessoa precisa ter acesso
        (autorização) para que lugares ela vai acessar dentro da empresa

### Criptografia simétrica e assimétrica:

- Simétrica: ambas as partes possuem uma mesma chave que vai tanto criptografar quanto descriptografar


- Assimétrica: são utilizadas duas chaves, uma para criptografar e outra para descriptografar
    - Propriedades: Têm duas funções: 
        - Criptografar a mensagem de forma que só o destinatário possa ler.
        - Assinar digitalmente a mensagem provando que foi você que escreveu aquela mensagem.   

### Autenticação HTTP:

- Basic: credenciais (usuário e senha).
Credencial (usuário:senha) -> Base64 -> HTTP Header Authorization

- APIKey: Criptografia simétrica.
Envia uma chave simétrica no: Header/ Body/ URL <- Não recomendado, pois a URL é logada em muitos sistemas.

- Bearer: tokens bearer (de portador) para acessar recursosprotegidos por OAuth 2.0. Criptografia assimétrica.
    Funciona por autenticações do OAuth 2.0.
    Alinhado com arquiteturas distribuídas como a de micro-seviços.
    Ao invés de confiar sua senha a cada API ou criar milhares de senhas para todas as APIs. Você confia em um único serviço que te
    fornece um token para acessar as outras APIs.
    Atualmente, a maioria dos sistemas usam o token no formato JWT.
    O token não é criptografado, mas sim assinado, o que garante que foi aquele servidor que o emitiu.
